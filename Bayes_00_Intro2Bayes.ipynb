# Most Basic Bayesian Interview Question

---

Are you familiar with Bayesian inference?

> Yes!

Well then you must know **the law of total probability**! What's that?

> What proportion of students will pass this class $\Pr(pass)$ if you know $\Pr(pass|quick)$ as well as the distribution of $quick$?

Now what proportion of students who will pass the class answer this question quickly?

> The structure of this question is **_the most basic Bayesian interview question_**, but it requires some assumptions...

# Bayesian Terminology 
---

## Bayes Theorem

\begin{align*}
\overset{posterior}{p(\theta|x)} &={} \frac{\overset{likelihood}{p(x|\theta)}\times \overset{prior}{p(\theta)}}{\overset{\text{evidence}}{p(x)}}\\
&={} \frac{{p(x, \theta)}}{\underset{\text{marginal likelihood}}{\displaystyle \int p(x|\theta)p(\theta) d\theta}} \textrm{ or }
\frac{{p(x, \theta)}}{\underset{\text{law of total probability}}{\displaystyle \sum_{k} p(x|\theta_k)p(\theta_k)}} =
\frac{\overset{joint}{p(x, \theta)}}{\underset{\text{normalizing constant}}{c}}
\end{align*}

### _The posterior is always proportional to the joint_

$$\large p(\theta|x) = (1/c)\times p(\theta,x) \propto{} p(\theta,x)$$

**_and the posterior is always a proper distribution if the prior is_**

# Example
---
\begin{align*}
\text{Likelihood} && p(x|\theta,n) &={} C^n_k\theta^{x}(1-\theta)^{n-x}\\
&&& \quad\; [\text{exponential family distribution if $n$ is known}]\\
&&& \quad\; [\text{What is this distribution? What is $C^n_k$?}]\\\\
\text{Prior} && p(\theta|\alpha,\beta) &={} \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\theta^{\alpha-1}(1-\theta)^{\beta-1} \\ 
&&& \quad\; [\text{$\alpha$ and $\beta$ are called }\textbf{hyperparameters}]\\
&&& \quad\; [\text{What is this distribution? What is $\Gamma(\cdot)$?}]\\\\
\text{Posterior} && p(\theta|x,n,\alpha,\beta) &={} \overset{\text{Bayes Theorem}}{\frac{p(x|\theta)p(\theta)}{p(x)}} = \frac{C^n_k\theta^{x}(1-\theta)^{n-x}  \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\theta^{\alpha-1}(1-\theta)^{\beta-1} }{p(x)} \\
&&& = \;\texttt{What distribution parameterized how?}
\end{align*}

# Answers

---

Q1:

Let A represent the event of passing the class.
Let B represent the event of answering the question quickly.

Then, we have,

\begin{align*}
&&P(A) = 0.9
&&P(B | A) = 0.6
&&P(B | A^{c}) = 0.3
&&P(A^{c}) = 1 - P(A) = 0.1
&&P(B) = P(B | A)P(A) + P(B | A^{c})P(A) = 0.57
\end{align*}

And so,

\begin{align*}
&&P(A | B) = \frac{P(B | A)P(A)}{P(B)} = \frac{0.6 \cdot 0.9}{0.57} = \frac{0.54}{0.57}
\end{align*}

Therefore, the proportion of students who pass the class provided that they answer the question quickly will be approximately $0.9474$.

Q2:

\begin{align*}
\text{Likelihood} && p(\textbf{x} | \boldsymbol{\theta}, n) = p(\textbf{x} = x_{1}, \ldots, x_{k} ; \boldsymbol{\theta} = \theta_{1}, \ldots, \theta_{k}, n) &={} 
\frac{\Gamma(\sum_{i=1}^{k} x_{i} + 1)}{\prod_{i=1}^k \Gamma(x_i + 1)} \prod_{i=1}^k \theta_{i}^{x_{i}}\\\\
\text{Prior} && p(\boldsymbol{\theta} | \boldsymbol{\alpha}) &={} 
p(\boldsymbol{\theta} = \theta_{1}, \ldots, \theta_{k}; \boldsymbol{\alpha} = \alpha_{1}, \ldots, \alpha_{k}) = \frac{\Gamma(\sum_{i=1}^{k} \alpha_{i})}{\prod_{i=1}^{k} \Gamma(\alpha_{i})} \prod_{i=1}^{k} \theta_{i}^{\alpha_i - 1}\\ 
\text{Posterior} && p(\boldsymbol{\theta} | \textbf{x}, \boldsymbol{\alpha}, n) &={} \overset{\text{Bayes Theorem}}{\frac{p(\textbf{x} | \boldsymbol{\theta})p(\boldsymbol{\theta})}{p(\textbf{x})}} = \frac{1}{p(\textbf{x})}(\frac{\Gamma(\sum_{i=1}^{k} x_{i} + 1)}{\prod_{i=1}^k \Gamma(x_i + 1)} \prod_{i=1}^k \theta_{i}^{x_{i}})(\frac{\Gamma(\sum_{i=1}^{k} \alpha_{i})}{\prod_{i=1}^{k} \Gamma(\alpha_{i})} \prod_{i=1}^{k} \theta_{i}^{\alpha_i - 1})\\
&&& \quad\quad\quad\quad\quad\; = \frac{1}{p(\textbf{x})}\frac{\Gamma(\sum_{i=1}^{k} x_{i} + 1)}{\prod_{i=1}^k \Gamma(x_i + 1)} \frac{\Gamma(\sum_{i=1}^{k} \alpha_{i})}{\prod_{i=1}^{k} \Gamma(\alpha_{i})}  \prod_{i=1}^k \theta_{i}^{x_{i}} \prod_{i=1}^{k} \theta_{i}^{\alpha_i - 1}\\
&&& \quad\quad\quad\quad\quad\; \propto \frac{\Gamma(\sum_{i=1}^{k} x_{i} + \alpha_{i} + 1)}{\prod_{i=1}^k \Gamma(x_i + \alpha_{i} + 1)} \prod_{i=1}^k \theta_{i}^{x_{i} + \alpha_{i} - 1}\\
&&& \text{Which is a Dirichlet distribution parameterized by} \hspace{0.25cm} \boldsymbol{\alpha} + \boldsymbol{x}
\end{align*}


Q3:

Let A, B, C be the three categories.

We shall take, as example, a multinomial distribution parameterized by $\theta = \{0.1, 0.1, 0.8\}$ and $n = 12$.

```python
import scipy
import numpy as np

import plotly.figure_factory as ff


# Helper function for generating grid points.
def grid_points(resolution: int=50) -> list[np.ndarray]:
    p1, p2 = np.mgrid[0:1:resolution * 1j, 0:1:resolution * 1j]
    p1, p2 = p1[p1 + p2 <= 1], p2[p1 + p2 <=1]

    p3 = 1 - p1 - p2

    return [p1, p2, p3]


# Helper function for computing the Multinomial probability per grid point.
def grid_probabilities(grid: list[np.ndarray]) -> np.ndarray:
    p1, p2, p3 = grid
    
    point_probabilities = np.zeros_like(p1)

    for i in range(len(p1)):
        point = np.round(n * np.array([p1[i], p2[i], p3[i]])).astype(int)

        point_probabilities[i] = scipy.stats.multinomial.pmf(point, n, probabilities)
    
    return point_probabilities


# Helper function for generating Dirichlet ternary plots.
def barycentric_coordinates(start: int, stop: int, resolution: int) -> np.ndarray:
    grid = np.linspace(start, stop, resolution)
    
    p1, p2 = grid.copy(), 1 - grid.copy()
        
    p1 = (p1[np.newaxis, :] * grid[:, np.newaxis]).flatten()
    p2 = (p2[np.newaxis, :] * grid[:, np.newaxis]).flatten()
    p3 = 1 - p1 - p2
    
    return (np.c_[p1, p2, p3]).T


# The number of trials.
n = 12

# The probability of a given sample being in each category A, B and C, respectively.
probabilities = [0.1, 0.1, 0.8]

# Create the multinomial distribution.
multinomial = scipy.stats.multinomial(n, probabilities)

# Sample from the multinomial distribution.
sample = multinomial.rvs(10)

# View the sample.
sample
```

```python
points = grid_points()

point_probabilities = grid_probabilities(points)

# Create the ternary contour plot for the Multinomial distribution.
fig = ff.create_ternary_contour(
    np.array(points),
    point_probabilities,
    pole_labels=['A', 'B', 'C'],
    interp_mode='cartesian',
    colorscale='Viridis',
    showscale=True,
    title='Multinomial Distribution'
)

# View the ternary plot for the Multinomial distribution.
fig.show('png')
```

```python
alpha = np.full(3, 5.0)

# Create the dirichlet prior distribution.
dirichlet_prior = scipy.stats.dirichlet(alpha=alpha)

# Create the dirichlet posterior distribution by updating the prior with a sampeld data point.
dirichlet_posterior = scipy.stats.dirichlet(alpha=(alpha + sample[0]))

# Generate barycentric coordinates for ternary plot.
coordinates = barycentric_coordinates(start=0, stop=1, resolution=100)
```

```python

# Create the ternary contour plot for the Dirichlet prior.
fig = ff.create_ternary_contour(
    coordinates, dirichlet_prior.pdf(coordinates),
    pole_labels=['a1', 'a2', 'a3'],
    interp_mode='cartesian',
    colorscale='Viridis',
    showscale=True,
    title='Dirichlet Prior'
)

# View the ternary plot for the Dirichlet prior.
fig.show('png')
```

```python
# Create the ternary contour plot for the Dirichlet posterior.
fig = ff.create_ternary_contour(
    coordinates, dirichlet_posterior.pdf(coordinates),
    pole_labels=['a1', 'a2', 'a3'],
    interp_mode='cartesian',
    colorscale='Viridis',
    showscale=True,
    title='Dirichlet Posterior'
)

# View the ternary plot for the Dirichlet posterior.
fig.show('png')
```
